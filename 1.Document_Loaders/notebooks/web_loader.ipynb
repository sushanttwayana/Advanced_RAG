### Recursive Loader

```python
from langchain_community.document_loaders import RecursiveUrlLoader
```

```python
# create the loader

base_url = "https://docs.langchain.com/oss/python/integrations/document_loaders"

loader = RecursiveUrlLoader(url=base_url,
                            max_depth=2)
```

```python
# load the documents

documents = loader.load()
```

```python
# number of documents

len(documents)
```

```python
### metadata fro first 10 documents

for i in range(10):
    doc = documents[i]
    print(doc.metadata, end="\n\n")
```

```python
documents[0].page_content[:500]
```

```python
documents[0].metadata
```

```python
# lazy load

documents_lazy_load = loader.lazy_load()
```

```python
documents_lazy_load
```

```python
counter = 0

for document in documents_lazy_load:
    # stop condition
    if counter == 20:
        break
    
    # incerement the counter
    counter += 1
    
    # print metadata and page content
    print(document.page_content[0:300])
    pp(document.metadata)
```
